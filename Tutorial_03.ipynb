```python
import numpy as np
import random

array_2x2 = np.array([[1, 2],
                      [3, 4]])
print("1. Array of size 2x2:\n", array_2x2)

array_random_3x4 = np.random.rand(3, 4)
print("\n2. Random array of size 3x4:\n", array_random_3x4)

array_ones = np.ones((2, 3))
array_zeros = np.zeros((2, 3))
print("\n3a. Array of ones (2x3):\n", array_ones)
print("\n3b. Array of zeroes (2x3):\n", array_zeros)

array_6x3 = np.random.randint(0, 10, size=(6, 3))
array_3x6_transposed = array_6x3.T
print("\n4a. Original random array (6x3):\n", array_6x3)
print("\n4b. Transposed array (3x6):\n", array_3x6_transposed)

n, p, m = 4, 3, 2

A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)

print("\n5a. Array A (4x3):\n", A)
print("\n5b. Array B (3x2):\n", B)
print("\n5c. Dot product (A dot B) result (4x2):\n", dot_product)
```
